```python
yrs = [2014,2015,2016,2017,2018,]
```

```python
yrs = np.arange(2012,2020)
```

```python
yrs
```

```python
apples = np.random.rand(8)
oranges = np.random.rand(8)
```

```python
plt.plot(yrs,apples);
```

```python
plt.plot(yrs,apples);
plt.plot(yrs,oranges);
plt.xlabel('Years');
plt.ylabel('Yield (tons per hect)');
```

# Chart Title and Legend

```python
plt.plot(yrs,apples);
plt.plot(yrs,oranges);
plt.xlabel('Years');
plt.ylabel('Yield (tons per hect)');

plt.title("Crop Yield");
plt.legend(['Apples','Oranges']);
```

```python
plt.plot(yrs,apples,'s-b');
plt.plot(yrs,oranges, 'o--r');
plt.xlabel('Years');
plt.ylabel('Yield (tons per hect)');

plt.title("Crop Yield");
plt.legend(['Apples','Oranges']);
```

```python
plt.figure(figsize=(10,8));
plt.plot(yrs,apples,'s-b');
plt.plot(yrs,oranges, 'o--r');
plt.xlabel('Years');
plt.ylabel('Yield (tons per hect)');

plt.title("Crop Yield");
plt.legend(['Apples','Oranges']);
```
