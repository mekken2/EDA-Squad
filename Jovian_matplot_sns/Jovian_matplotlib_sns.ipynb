# Adding Hues

```python
plt.figure(figsize=(10,8));
plt.title('Sepal Dimensions')
sns.scatterplot(flowers_df.sepal_length, flowers_df.sepal_width,hue=flowers_df.species,s=100);
```

# Plotting using Pandas DF

```python
plt.title('Sepal Dimensions')
sns.scatterplot('sepal_length','sepal_width',hue='species',s=100,data=flowers_df);
```

# Histogram

```python
flowers_df = sns.load_dataset("iris")
```

```python
flowers_df.sepal_width
```

```python
flowers_df.describe()
```

```python
plt.title("Distribution of Sepal Width")
plt.hist(flowers_df.sepal_width);
```

```python
plt.hist(flowers_df.sepal_width,bins=5);
```

```python
plt.hist(flowers_df.sepal_width,bins=np.arange(2,5,0.25));
```

# Multiple Histograms

```python
flowers_df
```

```python
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df = flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']
```

```python
plt.hist(setosa_df.sepal_width, alpha=0.4, bins=np.arange(2,5,0.25));
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins=np.arange(2,5,0.25));

plt.legend(['setosa','versicolor'])
```

```python
plt.hist([setosa_df.sepal_width,versicolor_df.sepal_width,virginica_df.sepal_width],bins=np.arange(2,5,0.25),stacked=True);

plt.legend(['setosa','versicolor','virginica']);
```

# Bar Plots with averages

```python
tips_df = sns.load_dataset("tips")
```

```python
tips_df
```

```python
bill_avg_df = tips_df.groupby('day')[['total_bill']].mean()
bill_avg_df
```

```python
plt.bar(bill_avg_df.index, bill_avg_df.total_bill);
```

```python
sns.barplot('day','total_bill',data=tips_df);
```

```python
sns.barplot('day','total_bill',hue='smoker',data=tips_df)
```

```python
sns.barplot('total_bill','day',hue='smoker',data=tips_df)
```

# Heatmap

```python
flights_df = sns.load_dataset('flights')
```

```python
flights_df
```

```python
plt.plot(flights_df.year,flights_df.passengers);
```

```python
flights_df = sns.load_dataset("flights").pivot('month','year','passengers')
flights_df
```

```python
plt.title("No of passengers")
sns.heatmap(flights_df)
```

```python
plt.figure(figsize=(10,8))
plt.title("No of passengers")
sns.heatmap(flights_df,fmt='d',annot=True,cmap="Blues")
```
