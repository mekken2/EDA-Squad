```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
import pandas as pd
```

# Basics

```python
yrs = np.arange(2012,2020)
```

```python
yrs
```

```python
apples = np.random.rand(8)
oranges = np.random.rand(8)
```

```python
plt.plot(yrs,apples);
```

```python
plt.plot(yrs,apples);
plt.plot(yrs,oranges);
plt.xlabel('Years');
plt.ylabel('Yield (tons per hect)');
```

# Chart Title and Legend

```python
plt.plot(yrs,apples);
plt.plot(yrs,oranges);
plt.xlabel('Years');
plt.ylabel('Yield (tons per hect)');

plt.title("Crop Yield");
plt.legend(['Apples','Oranges']);
```

```python
plt.plot(yrs,apples,'s-b');
plt.plot(yrs,oranges, 'o--r');
plt.xlabel('Years');
plt.ylabel('Yield (tons per hect)');

plt.title("Crop Yield");
plt.legend(['Apples','Oranges']);
```

```python
plt.figure(figsize=(10,8));
plt.plot(yrs,apples,'s-b');
plt.plot(yrs,oranges, 'o--r');
plt.xlabel('Years');
plt.ylabel('Yield (tons per hect)');

plt.title("Crop Yield");
plt.legend(['Apples','Oranges']);
```

# Scatter Plot

```python
flowers_df = sns.load_dataset("iris")
```

```python
flowers_df
```

```python
flowers_df.columns
```

```python
flowers_df.species.unique()
```

```python
plt.plot(flowers_df.sepal_length, flowers_df.sepal_width);
```

```python
sns.scatterplot(flowers_df.sepal_length, flowers_df.sepal_width);
```

# Adding Hues

```python
plt.figure(figsize=(10,8));
plt.title('Sepal Dimensions')
sns.scatterplot(flowers_df.sepal_length, flowers_df.sepal_width,hue=flowers_df.species,s=100);
```

# Plotting using Pandas DF

```python
plt.title('Sepal Dimensions')
sns.scatterplot('sepal_length','sepal_width',hue='species',s=100,data=flowers_df)
```
