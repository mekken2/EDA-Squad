# Aggregate Stats (groupby)

```python
df = pd.read_csv('poke_modified.csv')
```

```python
df
```

```python
df.groupby(['Type 1']).mean()
```

```python
df.groupby(['Type 1']).mean().sort_values('Attack',ascending=False)
```

```python
df
```

```python
df.groupby(['Type 2']).mean().sort_values('Defense',ascending=False)
```

```python
df.groupby(['Type 1']).count()
```

```python
df['count'] = 1
df.groupby(['Type 1']).count()['count']
```

```python
df.groupby(['Type 1','Type 2']).count()['count']
```
